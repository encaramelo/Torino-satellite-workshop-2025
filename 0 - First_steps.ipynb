## Linking Google Colab to your GitHub page or Google Drive

**You will nedd a Gmail account**

0 - Download the Folder that I provided with the email, and upload it to your own GitHub repository or Google Drive.

1 - Go to https://colab.research.google.com/

2 - Start session with your gmail credentials.

3 - Go to **File**, Open notebook

You can choose to take a notebook from **Google Drive** or **GitHub**, among others. Select the location where you stored the files in step 0.

4 - If using GitHub: Copy your personal GitHub repository address, and paste it in the URL address space. You will be able to see all your files there.

5 - Open **ee_api_colab_setup.ipynb**
